## Customer Analysis

## Strategic Recommendations

---
* Which region should receive more investment?
* Which product category should be promoted?
* Which customer segment should be targeted for growth?
* What actions would you recommend to increase profitability?

```python
# Which region should receive more investment?
df.groupby("Region")[["Revenue","Profit"]].sum()
```

```python
df.groupby("Region")[["Profit"]].mean()
```

```python
df.groupby("Region")[["Revenue","Profit","Cost"]].sum()
```

```python
df['profit_margin'] = (df['Profit'] / df['Revenue']) * 100
```

```python
df['profit_margin'].describe()
```

```python
df.groupby('Region')['profit_margin'].mean().sort_values(ascending=False)
```

```python
# Which product category should be promoted?
df.groupby("Product_Category")["Profit"].sum()
```

```python
df.head(1)
```

```python
# Which customer segment should be targeted for growth?
df.groupby("Customer_Segment")["Profit"].sum()
```
